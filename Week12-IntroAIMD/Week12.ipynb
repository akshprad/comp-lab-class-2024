```python
import MDAnalysis as mda
import numpy as np
import matplotlib.pyplot as plt
from MDAnalysis.analysis import rdf

# Load the .xyz file with MDAnalysis
u = mda.Universe('/home/ap8064/comp-lab-class/comp-lab-class-2024/Week12-IntroAIMD/Runs/Proton/proton.xyz')  # Replace with your .xyz file

# Manually define the box dimensions from the water.inp file
# Box dimensions: cubic box with side length 12.42 Å
u.dimensions = [12.42, 12.42, 12.42, 90.0, 90.0, 90.0]  # cubic box with 12.42 Å sides

# Select oxygen and hydrogen atoms
oxygen_atoms = u.select_atoms('name O')  # Select oxygen atoms
hydrogen_atoms = u.select_atoms('name H')  # Select hydrogen atoms

# Define the maximum radius (r_max) and bin width (bin_width) for the RDF calculation
r_max = 10.0  # Maximum radius for RDF (in Å)
bin_width = 0.1  # Bin width for RDF (in Å)

# Compute the RDF between oxygen and hydrogen atoms using InterRDF
rdf_analysis = rdf.InterRDF(oxygen_atoms, hydrogen_atoms, nbins=int(r_max/bin_width), range=(0.0, r_max))
rdf_analysis.run()

# Plot the radial distribution function g(r)
plt.plot(rdf_analysis.bins, rdf_analysis.rdf)
plt.xlabel('r (Å)')
plt.ylabel('g(r)')
plt.title('Radial Distribution Function g(r) for Oxygen-Hydrogen in Proton System')
plt.grid(True)
plt.show()
```
